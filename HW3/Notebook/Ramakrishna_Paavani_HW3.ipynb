```python
root_dir = '../Data/AReM/'
activity_folders = ['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
test_files = [
    '../Data/AReM/bending1\dataset1.csv', '../Data/AReM/bending1\dataset2.csv',
    '../Data/AReM/bending2\dataset1.csv', '../Data/AReM/bending2\dataset2.csv', 
    '../Data/AReM/cycling\dataset1.csv', '../Data/AReM/cycling\dataset2.csv', '../Data/AReM/cycling\dataset3.csv', 
    '../Data/AReM/lying\dataset1.csv', '../Data/AReM/lying\dataset2.csv', '../Data/AReM/lying\dataset3.csv', 
    '../Data/AReM/sitting\dataset1.csv', '../Data/AReM/sitting\dataset2.csv', '../Data/AReM/sitting\dataset3.csv',
    '../Data/AReM/standing\dataset1.csv', '../Data/AReM/standing\dataset2.csv', '../Data/AReM/standing\dataset3.csv',
    '../Data/AReM/walking\dataset1.csv', '../Data/AReM/walking\dataset2.csv', '../Data/AReM/walking\dataset3.csv'
]
```

```python
training_data = []
testing_data = []
training_labels = []
testing_labels = []
for activity in activity_folders:
    path_pattern = root_dir + activity + '/*.csv'
    for file_path in glob.glob(path_pattern):
        try:
            data_frame = pd.read_csv(file_path, skiprows=4, error_bad_lines=False)
        except Exception as e:
            print(f"Error reading file {file_path}: {e}")
            continue
        if file_path in test_files:
            testing_data.append(data_frame)
            testing_labels.append(activity)
        else:
            training_data.append(data_frame)
            training_labels.append(activity)
print("\n--- Test Data Overview ---")
for i, df in enumerate(testing_data):
    print(f"Activity: {testing_labels[i]} (Test Set)")
    print(df)
    print("\n")
print("\n--- Training Data Overview ---")
for i, df in enumerate(training_data):
    print(f"Activity: {training_labels[i]} (Training Set)")
    print(df)
    print("\n")
```

# (c) Feature Extraction: Classification of time series usually needs extracting features from them. In this problem, we focus on time-domain features.

i. Research what types of time-domain features are usually used in time series classification and list them (examples are minimum, maximum, mean, etc).

ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them directly. Where, for example, 1st quart6 means the first quartile of the sixth time series in each of the 88 instances.

```python
def extract_features(df):
    features = {}
    
    data_columns = df.columns[1:]
    for column in data_columns:
        series = pd.to_numeric(df[column], errors='coerce').dropna()
        if series.empty:
            stats = [np.nan] * 7
        else:
            stats = [
                series.min(),
                series.max(),
                series.mean(),
                series.median(),
                series.std(),
                series.quantile(0.25),
                series.quantile(0.75)
            ]
        features.update({
            f"{column}_Min": stats[0],
            f"{column}_Max": stats[1],
            f"{column}_Mean": stats[2],
            f"{column}_Median": stats[3],
            f"{column}_Std": stats[4],
            f"{column}_Q1": stats[5],
            f"{column}_Q3": stats[6],
        })
    return features

training_features = [extract_features(df) for df in training_data if not df.empty]
train_df = pd.DataFrame(training_features)

print("\n--- Extracted Features ---")
print(train_df.head())
```

iii. Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Pythons bootstrapped or any other method to build a 90% bootstrap confidence interval for the standard deviation of each feature.

```python
confidence_level = 0.90

table_data_bootstrap = []
features_data = train_df.columns
std_devs = train_df.std(ddof=1).values

for p in range(len(features_data)):
    samples = np.array(train_df[features_data[p]].dropna())
    boot_result = bs.bootstrap(samples, stat_func=bs_stats.std, alpha=0.1)
    table_data_bootstrap.append([
        features_data[p],
        std_devs[p],
        boot_result.lower_bound,
        boot_result.upper_bound
    ])

table_header_bootstrap = ['FEATURE NAME', 'STANDARD DEVIATION', 'LOWER BOUND', 'UPPER BOUND']
new_table_bootstrap = BeautifulTable(max_width=100)
new_table_bootstrap.set_style(BeautifulTable.STYLE_SEPARATED)
new_table_bootstrap.column_headers = table_header_bootstrap

for row in table_data_bootstrap:
    new_table_bootstrap.append_row(row)
print("\nStandard Deviation along with 90% Confidence Interval of Time Domain Features:")
print(new_table_bootstrap)
```

```python
test_features = []
for df in testing_data:
    features = extract_features(df)
    test_features.append(features)

test_df = pd.DataFrame(test_features)

table_data_bootstrap_test = []
features_data_test = test_df.columns
std_devs_test = test_df.std(ddof=1).values

for p in range(len(features_data_test)):
    samples = np.array(test_df[features_data_test[p]].dropna())
    boot_result = bs.bootstrap(samples, stat_func=bs_stats.std, alpha=0.1)
    table_data_bootstrap_test.append([
        features_data_test[p],
        std_devs_test[p],
        boot_result.lower_bound,
        boot_result.upper_bound
    ])

table_header_bootstrap_test = ['FEATURE NAME', 'STANDARD DEVIATION', 'LOWER BOUND', 'UPPER BOUND']
new_table_bootstrap_test = BeautifulTable(max_width=100)
new_table_bootstrap_test.set_style(BeautifulTable.STYLE_SEPARATED)
new_table_bootstrap_test.column_headers = table_header_bootstrap_test

for row in table_data_bootstrap_test:
    new_table_bootstrap_test.append_row(row)

print("\nStandard Deviation along with 90% Confidence Interval of Time Domain Features (Testing Data):")
print(new_table_bootstrap_test)
```

iv. Use your judgement to select the three most important time-domain features (one option may be min, mean, and max).

When tackling this regression problem, it's crucial to focus on time-domain features that offer the clearest separation and distinct distributions. Features that exhibit strong delineation between different data points will be the most valuable, as they can help improve model performance and interpretability.

Here are the three features that stand out as the most important:

- Maximum and Minimum Values (Extremes)
These features define the range of the data and help identify outliers or significant deviations. Maximum values capture peaks, while minimum values highlight the lowest pointsboth of which can significantly influence a datasets behavior. In many real-world scenarios, extreme values often indicate important events or patterns that shouldnt be overlooked.

- Mean (Central Tendency)
The mean represents the average value, providing a sense of where most data points are concentrated. Understanding the mean helps paint a broader picture of the dataset, offering insights into its overall trend. Especially in regression, knowing the average behavior of a variable is crucial for making accurate predictions.

- Standard Deviation (Variability & Dispersion)
Standard deviation measures how spread out the data points are around the mean. A higher standard deviation signals greater variability, meaning data points fluctuate more significantly, while a lower standard deviation suggests a more stable trend.
This feature is essential for understanding the consistency of the datawhether its tightly clustered or scattered across a wide range.

By focusing on extremes (max/min), central tendency (mean), and variability (std deviation), we gain a well-rounded understanding of the dataset. These features not only help in regression modeling but also provide interpretable insights that can be used for decision-making.
