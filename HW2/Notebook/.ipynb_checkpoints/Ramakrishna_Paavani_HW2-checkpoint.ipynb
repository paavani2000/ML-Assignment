iii. What are the mean, the median, range, first and third quartiles, and interquartile ranges of each of the variables in the dataset? Summarize them in a table.

```python
values = []
for column in df.columns:
    mean = df[column].mean()
    median = df[column].median()
    range_val = df[column].max() - df[column].min()
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    
    column_stats = {
        'Variable': column,
        'Mean': mean,
        'Median': median,
        'Range': range_val,
        'First Quartile': q1,
        'Third Quartile': q3,
        'Interquartile Range (IQR)': iqr
    }
    values.append(column_stats)
    
    table = pd.DataFrame(values)
table
```

(c)For each predictor, fit a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions. Are there any outliers that you would like to remove from your data for each of these regression tasks?

```python
def custom_function(x_column):
    threshold = 4/len(df)
    
    y = df['PE'].to_numpy()
    x = df[x_column].to_numpy()
    X = sm.add_constant(x)
    
    simple_reg_model = sm.OLS(y, X).fit()
    print("Summary of the model\n",simple_reg_model.summary())
    
    influence = simple_reg_model.get_influence()
    cooks = influence.cooks_distance
    outliers_idx = np.where(cooks[0] > threshold)
    
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', alpha=0.6, label='Data Points')
    plt.plot(x, simple_reg_model.predict(X), color='red', label='Regression Line')
    if len(outliers_idx[0]) > 0:
        plt.scatter(x[outliers_idx], y[outliers_idx], color='orange', label='Outliers', zorder=5)

    plt.title(f"Simple Linear Regression for {x_column}")
    plt.xlabel(x_column)
    plt.ylabel("PE")
    plt.legend()
    plt.show()
```

```python
independent_columns = ['AT','V','AP','RH']
for column in independent_columns:
    custom_function(column)
```

# Observations
1. Temperature (AT): The p-value for AT is 0, indicating a statistically significant relationship with PE.
2. Exhaust Vacuum (V): Similarly, V shows a significant association with PE, with a p-value of 0.
3. Ambient Pressure (AP): AP has a significant positive relationship with PE, with a p-value of 0.
4. Relative Humidity (RH): RH also demonstrates statistical significance with a p-value of 0.

AT and V are the strongest predictors of PE, with all four variables (AT, V, AP, RH) showing statistically significant relationships. Outliers, highlighted in orange on the plot, should be removed for more accurate results.

(d)Fit a multiple regression model to predict the response using all of the predictors.Describe your results. For which predictors can we reject the null hypothesis  H0 :Î²j = 0?

```python
X = df[['AT', 'V', 'AP', 'RH']] 
y = df['PE']   
X = sm.add_constant(X)
multi_reg_model = sm.OLS(y, X).fit()
print(multi_reg_model.summary())
```

(e) How do your results from 1c compare to your results from 1d? Create a plot displaying the univariate regression coefficients from 1c on the x-axis, and the multiple regression coefficients from 1d on the y-axis. That is, each predictor is displayed as a single point in the plot. Its coefficient in a simple linear regression model is shown on the x-axis, and its coefficient estimate in the multiple linear regression model is shown on the y-axis.

```python
simple_reg_coefs = []
multi_reg_coefs = []
y = df['PE']

for predictor in independent_columns:
    X_simple = sm.add_constant(df[predictor])
    simple_reg_model = sm.OLS(y, X_simple).fit()
    simple_reg_coefs.append(simple_reg_model.params[predictor])
    
X_multi = sm.add_constant(df[independent_columns])
multi_reg_model = sm.OLS(y, X_multi).fit()
multi_reg_coefs = multi_reg_model.params[independent_columns].tolist()
    
coefficients_df = pd.DataFrame({
    'Predictor': independent_columns,
    'Simple Regression Coef': simple_reg_coefs,
    'Multiple Regression Coef': multi_reg_coefs
})

plt.figure(figsize=(10, 6))
plt.scatter(coefficients_df['Simple Regression Coef'], coefficients_df['Multiple Regression Coef'], color='blue')
plt.title('Comparison of Simple vs Multiple Regression Coefficients')
plt.xlabel('Simple Regression Coefficients')
plt.ylabel('Multiple Regression Coefficients')

for i, txt in enumerate(coefficients_df['Predictor']):
    plt.annotate(txt, (coefficients_df['Simple Regression Coef'][i], coefficients_df['Multiple Regression Coef'][i]))

plt.show()
```

(f)Is there evidence of nonlinear association between any of the predictors and the response? To answer this question, for each predictor X, fit a model of the form Y =Î²0 +Î²1X+Î²2X2 +Î²3X3 +Îµ

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
def custom_function_non_linear(x_column):
    
    y = df['PE'].to_numpy()
    x = df[x_column].to_numpy()
    poly = PolynomialFeatures(degree=3, include_bias=False)
    X_poly = poly.fit_transform(x.reshape(-1, 1))
    
    # Fit the polynomial regression model
    model = sm.OLS(y, sm.add_constant(X_poly)).fit()
    print(model.summary())  # Print model summary
    
    # Generate predictions for plotting
    x_range = np.linspace(min(x), max(x), 100).reshape(-1, 1)
    x_range_poly = poly.transform(x_range)
    y_pred = model.predict(sm.add_constant(x_range_poly))
    
    # Plot the data and the fitted curve
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', alpha=0.6, label='Data Points')
    plt.plot(x_range, y_pred, color='red', label='Fitted Curve')
    
    plt.title(f"Polynomial Regression for {x_column}")
    plt.xlabel(x_column)
    plt.ylabel("Response")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
independent_columns = ['AT','V','AP','RH']
for column in independent_columns:
    custom_function_non_linear(column)
```

# Observation
We observe that p value is very small for features AT, AP, and RH providing evidence to reject the null hypothesis. This means these predictors and the response variables are associated non-linearly.But for the predictor V, the p-values with AP are not negligible. Notably, the interaction term V * AP shows a large p-value. This means that there is no non-linear association between the predictor V and the response variable.

(g) Is there evidence of association of interactions of predictors with the response? To answer this question, run a full linear regression model with all pairwise interaction terms and state whether any interaction terms are statistically significant.

```python
import statsmodels.formula.api as smf
pairwise_inter = 'PE ~ AT + V + AP + RH + AT*V + AT*AP + AT*RH + V*AP + V*RH + AP*RH'

model = smf.ols(formula = pairwise_inter, data = df)
model_pairwise = model.fit()
print("Model Summary:")
print(model_pairwise.summary())
```

# Observation:
We can see that significant interaction terms with very small p values less than 0.05 are

AT * V, AT * RH, V * AP, AP * RH
These interaction terms are statistically significant.

(h)Can you improve your model using possible interaction terms or nonlinear associations between the predictors and response? Train the regression model on a randomly selected 70% subset of the data with all predictors.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X= df[independent_columns] 
y=df['PE'] 
X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=104,  test_size=0.3,  shuffle=True)
X_train_c = sm.add_constant(X_train)
new_model = sm.OLS(y_train, X_train_c).fit()
print(new_model.summary())
y_pred = new_model.predict(X_train_c)
mse = mean_squared_error(y_train, y_pred)
print(f"Train MSE (Baseline Model): {mse}")
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
extended_model = sm.OLS(y_train, sm.add_constant(X_train_poly)).fit()
print("P-values:\n", extended_model.pvalues)
significant_columns = [i for i, p in enumerate(extended_model.pvalues) if p < 0.05]
print(f"Shape of X_train_poly: {X_train_poly.shape}")
print(f"Significant columns: {significant_columns}")

if 0 in significant_columns:
    significant_columns.remove(0)
X_train_poly_sig = X_train_poly[:, [i-1 for i in significant_columns]]  
X_train_poly_sig_const = sm.add_constant(X_train_poly_sig) 
adv_m = sm.OLS(y_train, X_train_poly_sig_const).fit()

print(adv_m.summary())

train_preds_extended = adv_m.predict(X_train_poly_sig_const)
train_mse_extended = mean_squared_error(y_train, train_preds_extended)
print(f"Train MSE (Extended Model): {train_mse_extended}")
```

```python
base_m = LinearRegression()
base_m.fit(X_train, y_train)
base_preds = base_m.predict(X_test)
test_mse_baseline = mean_squared_error(y_test, base_preds)
print("Test baseline mean squared error:",test_mse_baseline)

poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

adv_m = LinearRegression()
adv_m.fit(X_train_poly, y_train)
test_preds_extended = adv_m.predict(X_test_poly)
test_mse_extended = mean_squared_error(y_test, test_preds_extended)
print("Test extended mean squared error:",test_mse_extended)
```

```python
import pandas as pd
mse_data = {
    'Model': ['Baseline', 'Extended'],
    'Train MSE': [train_mse_baseline, train_mse_extended],
    'Test MSE': [test_mse_baseline, test_mse_extended]
}
mse_df = pd.DataFrame(mse_data)
print(mse_df)
```
